```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from google.colab import files
uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded)))
df.head()
```

```python

df_str = df.astype(str)

df['combined'] = df_str.agg(' '.join, axis=1)

df['combined'] = df['combined'].str.lower()


df[['combined']].head()
```

```python

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['combined'])
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend_restaurants(user_input, df, tfidf_matrix, vectorizer, top_n=5):
    user_input = user_input.lower()
    user_vec = vectorizer.transform([user_input])

    user_sim = cosine_similarity(user_vec, tfidf_matrix)

    sim_scores = list(enumerate(user_sim[0]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = [x for x in sim_scores if x[1] < 1.0]
    top_indices = [i for i, score in sim_scores[:top_n]]
    return df.iloc[top_indices]
```

```python

recommendations = recommend_restaurants("600 5.99", df, tfidf_matrix, vectorizer)
recommendations
```

```python

recommendations_coffee_coding = recommend_restaurants("600 8", df, tfidf_matrix, vectorizer)
print("Recommendations for '600 8':")
display(recommendations_coffee_coding)


recommendations_distractions_sleep = recommend_restaurants("1 8", df, tfidf_matrix, vectorizer)
print("\nRecommendations for '1 8':")
display(recommendations_distractions_sleep)
```
